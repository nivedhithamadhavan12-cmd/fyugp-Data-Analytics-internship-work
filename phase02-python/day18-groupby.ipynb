```python
import pandas as pd
df=pd.read_csv("SALE.csv")
df.head()
```

```python
#What is the total of a numeric column for each category
group_total=df.groupby("Category")["Sales"].sum()
group_total
```

```python
# What is the average for each category
group_average=df.groupby("Category")["Discount"].mean()
group_average
```

```python
#How many rows belong to each category (use .count() or value_counts())
value_count=df["Category"].value_counts()
value_count
```
