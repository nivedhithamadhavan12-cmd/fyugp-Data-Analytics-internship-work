```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("SALE.csv")
df
df.dtypes
```

```python
c_sales = df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(8,4))
plt.bar(c_sales.index, c_sales.values, color="orange")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Total sales by category")
plt.show()
```

```python
# This chart shows how total sales are distributed across different categories.
```

```python
plt.figure(figsize=(8,4))
#plt.plot(c_sales.index, c_sales.values, color="orange")
plt.plot(df["Sales"],color="red")
plt.title("Sales Trend")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.show()
```

```python
# This chart shows how sales values change across the dataset.
```

```python
plt.figure(figsize=(8,4))
plt.hist(df["Sales"], bins=20,color="green")
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# This chart shows how sales values are distributed and where most values fall.
```

```python
df12=pd.read_csv("day12datasetclean.csv", encoding="latin1")
#df12['Sales'] = df12['Sales'].astype(int)
df12["Sales"]=df12["Sales"].astype(str).str.replace(r'[^0-9.]',"",regex=True)
df12['Sales'] = pd.to_numeric(df12['Sales'], errors='coerce')
```

```python
category_sales = df12.groupby("Category")["Sales"].sum()
```

```python
#plt.bar(df12["Category"],df12["Discount"], color="orange")
plt.bar(category_sales.index,category_sales.values,color="yellow")
plt.title("Total Sales by Category")
plt.xlabel("category")
plt.ylabel("Sales")

plt.show()
```

```python
# This chart recreates the Power BI visualization and shows similar sales patterns by category.
```
