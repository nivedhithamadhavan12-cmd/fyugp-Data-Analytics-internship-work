```python
import pandas as pd
df=pd.read_csv("messy_sales_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# This shows how many missing values exist in each column 
```

```python
df=df.dropna(subset=["Age"])
```

```python
# Drop rows where a key column is missing or null  because they cannot be reliably used in analysis.
```

```python
df["Gender"]=df["Gender"].fillna("Unknown")
df["Gender"]
```

```python
# Missing values were filled to avoid losing data 
```

```python
df=df.drop_duplicates()
df
```

```python
## Duplicate rows were removed to prevent double counting and incorrect analysis.
```

```python
df.dtypes
```

```python
df["Sales"]=pd.to_numeric(df["Sales"],errors="coerce")
df["OrderDate"]=pd.to_datetime(df["OrderDate"],errors="coerce")
df.dtypes
```

```python
# Data type was corrected to ensure proper calculations and analysis
```
