```python
#celsius_to_fahrenheit
def celsius_to_fahrenheit(n):
    f=(n*(9/5))+32
    print("Celsius to fahrenheit:",f)

celsius_to_fahrenheit(20)
celsius_to_fahrenheit(30)
celsius_to_fahrenheit(40)
```

```python
# get_grade
def  get_grade(n):
    if n>=90:
        return("A")
    elif n>=75:
        return("B")
    elif n>=50:
        return "C"
    else:
        return "F"
        
```

```python
get_grade(20)
```
