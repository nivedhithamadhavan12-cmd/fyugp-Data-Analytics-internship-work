```python
score=[45,67,80,30,50,71,31,95,55,40]
print(score)
```

```python
for s in score:
    if(s>=50):
        print(s,":Pass")
    else:
        print(s,":Fail")
```

```python
pass_count=0
fail_count=0
for s in score:
    if(s>=50):
        pass_count+=1
    else:
        fail_count+=1        
```

```python
print(pass_count ,"students passed out of 10")
print(fail_count,"students fail ")
```
